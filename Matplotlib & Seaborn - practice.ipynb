```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Matplotlib

### Scatterplot

```python
#Scatterplot

plt.scatter(x,y,c='black')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('X-Y plot')
plt.show()

#to save the figure in specified directory
#plt.savefig('Test.png')
```

#### Scatterplot - plotline
- if instead of points, i need the line plot.

```python
y=x*x
plt.plot(x,y,'r.',linestyle='dashed', linewidth = 2,markersize=12)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2-D diagram X-Y Plot')
plt.show()
```

### Creating Sub-plots

```python
# plt.subplot(norw =, ncol= , position of plot)

plt.subplot(2,2,1)
plt.plot(x,y,'k+')
plt.subplot(2,2,2)
plt.plot(x,y,'r+--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.show()
```

```python
# computing x & y co-ordinates for points on sine & cosine curve

x = np.arange(0,5*np.pi,0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)

# Set up a subplot grid that has height 2 and width 1( 2 rows , 1 column) & position as 1
plt.subplot(2,1,1)

plt.plot(x,y_sin,'k.--')
plt.title('Sine')

# Set up a subplot grid that has height 2 and width 1( 2 rows , 1 column) & position as 2
plt.subplot(2,1,2)

plt.plot(x,y_cos, 'r*--')
plt.title('Cosine')

plt.show()
```

### Barplot

```python
x1 = [2,8,10]
x2 = [11,16,19]

y1 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1,color='b')
plt.bar(x2,y2,color='green')
plt.title('Bar-plots')

plt.show()
```

### Histogram - gives the counts ( in bins)

```python
a = np.array([22,87,34,55,67,45,66,8,78,39,98])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

### Boxplot - helpful in statistical inference. (percentiles)

```python
data = [np.random.normal(loc=0,scale=std,size=100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
plt.show()
```

```python
data
```

### Piechart

```python
labels = 'Python', 'R', 'Machine_Learning','MLOPS', 'Deep learning'
sizes = [200, 300, 450, 300, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.4,0,0,0,0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

### Distplot 
- checks the distribution of column features (histogram)

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.displot(df['tip'],kde=True,bins=10)
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```

## Correlation plot

```python
df.dtypes
```

```python
numeric_columns = df.select_dtypes(include=['number'])
numeric_columns.corr()
```

```python
sns.heatmap(numeric_columns.corr())
```

## JoinPlot : 
    - Joinplot is used for finding relationship between 2 numeric columns.
        - The central chart displayes correlation.
        - It is usually a scatterplot , a hexbin plot, a 2D histogram or a 2D density plot

### Bivariate analysis:

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
# With regression line
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pairplot
- A pairplot is a scatterplot of each variable with each other variable available in the data.

```python
sns.pairplot(df)
```

```python
# also looking at category wise distribution - add 'hue=column_name'
sns.pairplot(df,hue='smoker')
```

### Violin-plot
- It helps us look at both the distribution of data in kernel density estimation & the box-plot.

```python
sns.violinplot(x='total_bill',y='day',hue='smoker',data=df, palette='rainbow')
```
